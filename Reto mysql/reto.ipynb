## Enumera el numero total de medallas de oro, plata y bronce ganadas por cada pais en relacion con cada juego olimpico

```python
q="""
SELECT
    Year,
    Season,
    NOC,
    SUM(CASE WHEN Medal = 'Gold' THEN 1 ELSE 0 END) AS Gold_Medals,
    SUM(CASE WHEN Medal = 'Silver' THEN 1 ELSE 0 END) AS Silver_Medals,
    SUM(CASE WHEN Medal = 'Bronze' THEN 1 ELSE 0 END) AS Bronze_Medals
FROM
    f
WHERE
    Medal IN ('Gold', 'Silver', 'Bronze')
GROUP BY
    Year,
    Season,
    NOC
ORDER BY
    Year,
    Season,
    NOC;

"""
medals_countries_year = sqldf(q)

medals_countries_year
```

## 15. Identifica que pais gano la mayoria de las medallas de oro, plata y bron en cada juego olimpico.

```python
q="""
WITH MedalCounts AS (
    SELECT
        Year,
        Season,
        NOC,
        COUNT(CASE WHEN Medal = 'Gold' THEN 1 END) AS Gold_Medals,
        COUNT(CASE WHEN Medal = 'Silver' THEN 1 END) AS Silver_Medals,
        COUNT(CASE WHEN Medal = 'Bronze' THEN 1 END) AS Bronze_Medals
    FROM
        f
    WHERE
        Medal IN ('Gold', 'Silver', 'Bronze')
    GROUP BY
        Year,
        Season,
        NOC
)
SELECT
    Year,
    Season,
    NOC,
    CASE
        WHEN Gold_Medals >= Silver_Medals AND Gold_Medals >= Bronze_Medals THEN 'Gold'
        WHEN Silver_Medals >= Gold_Medals AND Silver_Medals >= Bronze_Medals THEN 'Silver'
        ELSE 'Bronze'
    END AS Most_Common_Medal,
    CASE
        WHEN Gold_Medals >= Silver_Medals AND Gold_Medals >= Bronze_Medals THEN Gold_Medals
        WHEN Silver_Medals >= Gold_Medals AND Silver_Medals >= Bronze_Medals THEN Silver_Medals
        ELSE Bronze_Medals
    END AS Num_Most_Common_Medal
FROM
    MedalCounts
ORDER BY
    Year,
    Season,
    NOC;

"""
medals_countries_olimpicgame = sqldf(q)

medals_countries_olimpicgame
```
